```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/content/netflix_titles_nov_2019.csv')
```

```python
df['director']= df['director'].fillna('Unknown')
df['cast']= df['cast'].fillna('Unknown')
df['country']= df['country'].fillna('Unknown')
df.dropna(subset=['date_added','rating'], inplace=True)
```

```python
df['date_added']= pd.to_datetime(df['date_added'].str.strip())
df['year_added']= df['date_added'].dt.year
```

```python
sns.countplot(x='type', data=df, palette='Reds')
plt.title('Content Type Distribution')
```

```python
content_by_year= df.groupby('year_added')['show_id'].count().reset_index()
sns.lineplot(x='year_added', y='show_id', data=content_by_year)
plt.title('Content Added to Netflix Over the Years')
```

```python
genres= df['listed_in'].str.split(',').explode()
top_genres= genres.value_counts().head(10)
top_genres.plot(kind='barh', color='darkred')
plt.title('Top 10 Genres on Netflix')
```
